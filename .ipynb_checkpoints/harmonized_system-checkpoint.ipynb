```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tflearn

from collections import Counter
```

```python
csv = pd.read_csv("harmonized_system_utf_8.csv", dtype="str")
```

```python
print(csv[:5])
```

```python
rows = csv[:10]
count_level_1 = Counter()
count_level_2 = Counter()
count_level_3 = Counter()
for i, row in enumerate(csv.values):
    order, level, short_code, long_code, long_parent, code, parent, desc = row
    
    # if short_code is less than 5 add a 0 in the front
    if(len(short_code)==5): 
        short_code = "0" + short_code
    
    count_level_1[short_code[0:2]] += 1
    count_level_2[short_code[2:4]] += 1
    count_level_3[short_code[4:6]] += 1

    # print(short_code, short_code[0:2], short_code[2:4], short_code[4:6], code, desc)
    
```

```python
def hot_encoder(labels, max_value):
    labels_hot_encoded = {}
    for label in labels:
        zeros_array = np.zeros(max_value)
        zeros_array[label - 1] = 1
        labels_hot_encoded[label] = zeros_array
    return labels_hot_encoded
```

```python
hot_encoder([1, 5, 3], 5)
```
